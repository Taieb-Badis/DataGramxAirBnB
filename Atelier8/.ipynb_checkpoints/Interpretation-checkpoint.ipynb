# Interprétation du modèle

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import shap

import xgboost as xgb

shap.initjs()
```

```python
import model_creator

booster, X_train, X_test, Y_train, Y_test = model_creator.create_model("../data/price_availability.csv", "../clean/listings_final.csv")
```

```python
explainer = shap.TreeExplainer(booster)
shap_values = explainer.shap_values(X_test, approximate=True)
index = 78
shap.force_plot(explainer.expected_value, shap_values[index,:], X_test.iloc[index,:])
```

```python
shap.force_plot(explainer.expected_value, shap_values, X_train)
```

```python
shap.dependence_plot("person_capacity", shap_values, X_test)
```

```python
shap.summary_plot(shap_values, X_test)
```

```python
shap.summary_plot(shap_values, X_test, plot_type="bar")
```
